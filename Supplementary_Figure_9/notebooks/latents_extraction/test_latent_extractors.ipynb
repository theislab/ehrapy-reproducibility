```python
import numpy as np
import pytorch_lightning as pl
import torch


def predict_latents(extractor, dataloader):

    # meant for predicting latents

    trainer = pl.Trainer(limit_predict_batches=2)
    lat = trainer.predict(model=extractor.model, dataloaders=dataloader)

    latents = torch.cat([el["latents"] for el in lat]).cpu().detach().numpy()

    gt = np.concatenate([el["gt"] for el in lat])
    pat_id = np.concatenate([el["id"] for el in lat])

    return latents, {"id": pat_id, "gt_class": gt}
```

```python
latents, info_dict = predict_latents(extractor, dataloader)
```
